# MOVIE RECOMMENDATION SYSTEM USING SURPRISE PACKAGE

```python
#Importing libraries
import pandas  as pd
import numpy as np
```

```python
#importing Dataframes
ratings = pd.read_csv("ratings.csv")
movies = pd.read_csv("movies.csv")
```

```python
ratings.head()
```

```python
movies.head()
```

```python
#Merging Two dataframes
df = pd.merge(ratings, movies, on = "movieId")
```

```python
df.head()
```

```python
#Checking NULLVALUES
df.isnull().sum()
```

```python
#Total rating counts 
movie_rating_count = df.groupby("title")["rating"].count().reset_index().rename({"rating": "total_rating_count"}, axis = 1)
```

```python
movie_rating_count.head()
```

```python
rating_with_totalRatingCount = df.merge(movie_rating_count, on = "title")
```

```python
rating_with_totalRatingCount.head()
```

```python
import seaborn as sns
sns.distplot(rating_with_totalRatingCount["total_rating_count"], kde = False)
```

```python
threshold = 50 #minimum threshold for total rating count per movie for popular movie
popular_movies= rating_with_totalRatingCount[rating_with_totalRatingCount["total_rating_count"]>=threshold]
```

```python
popular_movies.head()
```

```python
popular_movies.shape
```

```python
#Creating Pivot Matrix
movie_features = popular_movies.pivot_table(index = "title", columns = "userId", values = "rating").fillna(0)
```

```python
movie_features
```

```python
#creatinng sparse compressed matrix by using csr_matrix
from scipy.sparse import csr_matrix
movie_features_matrix = csr_matrix(movie_features.values)
```

```python
#importing Nearest Neighbors Library
from sklearn.neighbors import NearestNeighbors
model_knn = NearestNeighbors(metric = "cosine",algorithm= "brute")
model_knn.fit(movie_features_matrix)
```

```python
#selecting a random index from the dataframe to find its nearest movies 
query = np.random.choice(movie_features.shape[0])
print(query)
distance, indeces = model_knn.kneighbors(movie_features.iloc[query,:].values.reshape(1,-1),n_neighbors= 10)
```

```python
distance
```

```python
indeces
```

```python
movie_features.index[indeces.flatten()]
```

```python
for i in range(0, len(distance.flatten())):
    if i == 0:
        print('Recommendations for {0}:\n'.format(movie_features.index[query]))
    else:
        print('{0}: {1}, with distance of {2}:'.format(i, movie_features.index[indeces.flatten()[i]], distance.flatten()[i]))
```

# USING SUPRISE PACKAGE TO PREDICT  THE MOVIE RATING


```python
#finding range of rating present in dataframe
lower_rating = df["rating"].min()
highest_rating = df["rating"].max()
print("Highest Rating:", highest_rating)
print("Lowest Rating:", lower_rating)
```

```python
#importing surprise package to find movie rating for the user for movie which he hasn't watched
from surprise import KNNWithMeans,Dataset,accuracy,Reader
from surprise.model_selection import train_test_split
```

```python
reader = Reader(rating_scale = (0.5,5))
data = Dataset.load_from_df(df[["userId","movieId","rating"]], reader)
train, test = train_test_split(data, test_size = 0.15)
```

```python
#fitting algorithm 
alg=KNNWithMeans(k=50,sim_options={"name":"cosine","user_based":True})
alg.fit(train)
# test_pred=recom.test(test)
# RMSE=accuracy.rmse(test_pred)
```

```python
#using user ID and movieid as input to predict rating for selected movie id
userid = str(input("Enter User Id:"))
movieid = int(input("Enter Movie Id:"))

pred = alg.predict(uid = userid, iid = movieid)
print("Predicted Rating of User Id {0} for Movie Id {1} is:".format(userid,movieid),pred.est) 
```

## Recommending a user to watch top 10 movies, which has not watched yet (as per his rating predictions).  By taking userID as a keyboard input. Fix the threshold rating as 2.5 

```python
#user id as input
userID = int(input("Enter User Id:"))
#getting list of all movie ids
iids = df["movieId"].unique()
#getting list of movies that userId 5 has rated
uids = df.loc[df["userId"]==userID, "movieId"]
#removing the movie id that userId 5 HAS RATED FROM THE LIST OF ALL MOVIE id(GETTING THE MOVIES LIST THAT USER 5 HAS NOT WATCHED)
iids_to_pred = np.setdiff1d(iids, uids)


#Preaparing a testset
testset = [[userID,movie_id,2] for movie_id in iids_to_pred]
prediction = alg.test(testset)

#Preparing a dataframe consisting of TOP 10 MOVIES with its Movie ID as Index and RATING as Values.
usid=[]
rate = []
for i in prediction:
    usid.append(i.iid)
    rate.append(i.est)
predDf = pd.DataFrame({"MovieID":usid,"Ratings":rate})
predDf=predDf.set_index("MovieID")
topten = predDf.sort_values(by="Ratings",ascending=False).head(10)
top_10_movie_id = pd.Series(topten.index)
top_10_movie_id

#Adding the Movie title to the above dataframe by using above top_10_movie_id dataframe and movies dataframe
top10_movies_recommended = movies[movies["movieId"].isin(top_10_movie_id)]["title"]
print("movies to be recommended are", sep= "\n" )
print(top10_movies_recommended)
 
```

# Displaying the MovieID, IMDB ID, Average User Rating (excluding predictions), genres and tag of all the Movies Recommende as a data frame.
